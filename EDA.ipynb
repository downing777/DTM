```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('产品数据.xlsx', sheet_name = '成人乳胶枕1')
df.head()
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
#成人乳胶枕
df['日期'] = pd.to_datetime(df['日期'])

# 创建图形对象
fig, ax1 = plt.subplots(figsize=(14, 7))

# 绘制直方图
ax1.bar(df['日期'],df['交易金额'], color='skyblue', edgecolor='black', alpha=0.7)
ax1.set_xlabel('Sales')
ax1.set_ylabel('Frequency')
ax1.set_title('Daily Sales Histogram and Sales Trend')

# 创建第二个y轴
ax2 = ax1.twinx()

# 插值以获得光滑曲线
x = df['日期'].astype('int64')  # 将日期转换为int64类型
y = df['交易金额'].values
x_smooth = np.linspace(x.min(), x.max(), 300)  # 创建更多的x点用于平滑曲线
spl = make_interp_spline(x, y, k=3)  # 使用三次B样条进行插值
y_smooth = spl(x_smooth)

# 将x_smooth转换回日期格式
x_smooth_dates = pd.to_datetime(x_smooth)

# 绘制光滑曲线
ax2.plot(x_smooth_dates, y_smooth, color='blue', lw=3)
ax2.set_ylabel('Sales')
ax2.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_excel('产品数据.xlsx', sheet_name = 'Sheet1')
df1 = df1.dropna()
df1 = df1[df1['行标签']<=12]
sales = df1[' 销量'].to_list()
dates = pd.date_range(start='2016-09', end='2023-12', freq='M').strftime('%Y-%m')



# 创建图形对象
fig, ax1 = plt.subplots(figsize=(14, 7))

# 绘制柱状图

ax1.bar(dates, sales, color='skyblue')
ax1.set_title('Monthly Sales Histogram and Sales Trend')
ax1.set_xlabel('Index')
ax1.set_ylabel('Values')

# 每三个月显示一个日期
xticks_positions = range(0, len(dates), 3)
xticks_labels = [dates[i] for i in xticks_positions]
ax1.set_xticks(xticks_positions)
ax1.set_xticklabels(xticks_labels, rotation=45, ha='right')

ax2 = ax1.twinx()
  
# 绘制曲线图
ax2.plot(dates, sales, marker='o', color='red')


plt.show()
```

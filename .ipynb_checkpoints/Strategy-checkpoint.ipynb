```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('./Dataset/latex_pillow.csv')
df1_normalized = pd.read_csv('./Dataset/latex_pillow_normalized.csv')
```

```python
# 2. 定义 LSTM 模型
import yaml
import torch
import joblib
import numpy as np
class SalesLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SalesLSTM, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)
    
    def forward(self, x, hidden):
        lstm_out, hidden = self.lstm(x, hidden)
        predictions = self.fc(lstm_out)
        return predictions, hidden



class SalesPredictor:
    def __init__(self, config_path):
        """
        初始化SalesPredictor类，自动加载训练好的模型和scalers。

        参数:
        config_path: 配置文件路径，包含模型路径和超参数信息。
        """
        # 加载配置文件
        with open(config_path, 'r') as file:
            config = yaml.safe_load(file)
        
        # 加载scalers
        self.feature_scaler = joblib.load(config['model']['feature_scaler_path'])
        self.target_scaler = joblib.load(config['model']['target_scaler_path'])

        # 初始化模型
        self.input_size = config['lstm_params']['input_size']
        self.hidden_size = config['lstm_params']['hidden_size']
        self.output_size = config['lstm_params']['output_size']

        self.model = SalesLSTM(input_size=self.input_size, hidden_size=self.hidden_size, output_size=self.output_size)
        self.init_hidden =  (torch.zeros(1, 1, self.hidden_size),torch.zeros(1, 1, self.hidden_size))

        # 加载训练好的模型参数
        self.model.load_state_dict(torch.load(config['model']['model_path']))
        self.model.eval()  # 设置为评估模式
    
    def predict_next(self, input_features, hidden = None):
        """
        根据输入的特征数据预测下个月的销量

        参数:
        input_data: 包含当月特征的列表或NumPy数组，形状为(1, input_size)，
                    例如[成本, 单价, 毛利率, 当月销量, ...]

        返回:
        预测的下个月销量
        """
        # 将输入数据归一化
        input_data_scaled = self.feature_scaler.transform(np.array(input_features).reshape(1, -1))

        # 将数据转换为PyTorch张量
        input_tensor = torch.tensor(input_data_scaled, dtype=torch.float32).unsqueeze(0)  # 形状 (1, 1, input_size)

        # 初始化隐藏状态
        if not hidden:
            hidden = self.init_hidden

        # 使用模型进行预测
        with torch.no_grad():
            prediction, updated_hidden = self.model(input_tensor, hidden)
    
        predicted_sales = prediction.numpy().flatten()[0]
        # 反归一化预测结果
        inverse_predicted_sales = self.target_scaler.inverse_transform([[predicted_sales]])[0][0]
        
        
        return predicted_sales, inverse_predicted_sales, updated_hidden
```

```python
# 提取特征和目标变量
features = ['Month','Sales_volume','Unit_price', 'Promotion_expense', 'ROI','Profit_rate','Unit_cost']# please follow the input order
target = 'Target'
```

```python
model = SalesPredictor('config.yaml')
# 按照版本切割数据
version_groups = df1.iloc[:-1].groupby('Version')

# 为每个版本的数据创建 DataLoader
dataloaders = []

for version, group in version_groups:
    X = group[features].values
    y = group[target].values
    
    # 创建 TensorDataset 并整合为 DataLoader
    dataset = (X, y)
    
    # 存储 DataLoader
    dataloaders.append(dataset)
```

```python
X = dataloaders[0][0]
X[0]
```

```python
# 预测并存储结果
predictions = []
targets = []
hidden_size = model.hidden_size

for dataset in dataloaders:
    X_train = dataset[0]
    y_train = dataset[1]
    hidden = (torch.zeros(1, 1, hidden_size),
              torch.zeros(1, 1, hidden_size))

    with torch.no_grad():
        pred_series = []
        for i in range(len(X_train)):
            x_input = X_train[i]  # 输入数据
            y_pred,y_inverse, hidden = model.predict_next(x_input, hidden)
            predictions.append(y_inverse)

    targets.extend(y_train)

# 可视化
plt.figure(figsize=(10, 5))
plt.plot(targets, label='Actual Sales')
plt.plot(predictions, label='Predicted Sales', linestyle='--')
plt.title('Comparison of Actual and Predicted Sales')
plt.xlabel('Time (Months)')
plt.ylabel('Sales Volume')
plt.legend()
plt.show()
```

```python
# 假设 df_normalized 是包含所有归一化后数据的 DataFrame
# 计算每个月份的平均特征值
monthly_averages = df1.groupby('Month').mean()

# 初始化最后一个已知数据点，假设 month_index 是月份在特征中的索引位置
month_index = 0
sales_index = 1
last_month_data = df1.iloc[-1][features].values
current_month = last_month_data[month_index]

input_features = last_month_data
future_predictions = []

hidden = (torch.zeros(1, 1, hidden_size),
              torch.zeros(1, 1, hidden_size))

for i in range(12):  # 预测未来12个月
    with torch.no_grad():
        # 模型预测
        _, sales_prediction, hidden = model.predict_next(input_features, hidden)
        
        future_predictions.append(sales_prediction)
        
        # 计算下一个月份
        next_month = (current_month + 1)%12
        if next_month == 0:
            next_month = 12

        input_features[month_index] = next_month  # 重新归一化月份
        
        # 更新销量
        input_features[sales_index] = sales_prediction
        
        # 更新其他特征为历史平均值
        for idx, feature in enumerate(features):
            if feature not in ['Month', 'Sales_volumn']:  # 假设月份和销量已处理
                input_features[idx] = monthly_averages.loc[next_month, feature]

# 可视化未来一年的销量预测
# 合并过去拟合数据和未来预测数据
combined_predictions = predictions + future_predictions

# 生成时间轴
months_past = list(range(1, len(targets) + 1))  # 过去数据的月份
months_combined = list(range(1, len(targets) + len(future_predictions) + 1))  # 组合后的时间轴

# 开始绘图
plt.figure(figsize=(10, 5))

# 绘制实际数据
plt.plot(months_past, targets, label='Actual Sales', color='blue')
# 绘制拟合和未来预测数据
plt.plot(months_combined, combined_predictions, label='Predicted Sales', linestyle='--', color='green')

# 添加标题和轴标签
plt.title('Comparison of Actual and Predicted Sales')
plt.xlabel('Time (Months)')
plt.ylabel('Sales Volume')

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
